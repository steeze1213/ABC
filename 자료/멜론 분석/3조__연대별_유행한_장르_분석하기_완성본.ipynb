##3.데이터 시각화 하기

### 1990년대 음악 장르 시각화하기

```python
color_map = {'발라드': 'springgreen',
             '댄스':'crimson',
             '랩/힙합':'navy',
             '랩':'navy',
             'R&B/Soul':'aqua',
             'R&B':'aqua',
             '인디음악':'darksalmon',
             '성인가요/트로트':'pink',
             '록/메탈': 'fuchisia',
             '록':'fuchisia',
             '트로트':'cadetblue',
             '포크/블루스':'orange'
             }

px.pie(song_dict_90, values='count', names='장르',color='장르',color_discrete_map=color_map,title='1990년대 유행한 음악 장르')
```

### 2000년대 음악 장르 시각화하기

```python
px.pie(song_dict_00, values='count', names='장르',color='장르',color_discrete_map=color_map,title='2000년대 유행한 음악 장르')
```

### 2010년대 음악 장르 시각화 하기

```python
px.pie(song_dict_10, values='count', names='장르',color='장르',color_discrete_map=color_map,title='2010년대 유행한 음악 장르')
```

```python
# 함수 정의: color_map을 사용하여 색상 리스트 생성
def get_colors(labels, color_map):
    return [color_map[label] for label in labels]
# labels 는 리스트 형태 // color_map 은 딕셔너리 형태
# color_map 의 색상을 키값으로 labels 안에서 뽑는다 -> 이를 리스트화 [] 시켜 return 한다 (반환)

# 각각의 파이 차트를 생성하고 색상 지정
colors_90 = get_colors(song_dict_90['장르'], color_map)
colors_00 = get_colors(song_dict_00['장르'], color_map)
colors_10 = get_colors(song_dict_10['장르'], color_map)

# make_subplots를 사용하여 틀 생성
fig = make_subplots(rows=1, cols=3, specs=[[{'type': 'domain'}, {'type': 'domain'}, {'type': 'domain'}]],
                    subplot_titles=["1990년대", "2000년대", "2010년대"])

# add_trace : 각 파이 차트를 fig 에 추가한다 // labels 는 요소(names) , values는 똑같이 값
fig.add_trace(go.Pie(labels=song_dict_90['장르'], values=song_dict_90['count'], marker=dict(colors=colors_90), name="1990s"), 1, 1)
fig.add_trace(go.Pie(labels=song_dict_00['장르'], values=song_dict_00['count'], marker=dict(colors=colors_00), name="2000s"), 1, 2)
fig.add_trace(go.Pie(labels=song_dict_10['장르'], values=song_dict_10['count'], marker=dict(colors=colors_10), name="2010s"), 1, 3)

# 전체 레이아웃 업데이트 - 제목 설정
fig.update_layout(
    title_text="10년 단위의 멜론 TOP50 음악 장르 양상",
)

fig.show()
```
