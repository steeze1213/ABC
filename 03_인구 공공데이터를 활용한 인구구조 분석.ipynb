## 2.신도림 지역의 인구 구조 시각화

```python
print('서울특별시 구로구 신도림동(1153051000)' == '신도림')

# in 연산자를 활용해서 이름 포함한 지역 찾기
print('서울특별시 구로구 신도림동(1153051000)' in '신도림')
print('신도림' in '서울특별시 구로구 신도림동(1153051000)')
```

```python
#인구 구조 결과를 담을 수 있는 리스트 선언
result = []

# 1) 파일 읽기
f = open('/content/drive/MyDrive/ABC/201902_201902_연령별인구현황_월간.csv')

data = csv.reader(f)

# 2) 데이터  확인
for row in data:
  # 3) 신도림 지역의 정보만 담기
  if '신도림' in row[0]:
    # 4) 0~99세 인구수 담기
    for i in row[3:]:
      result.append(int(i.replace(',','')))

# 5) 연령별 인구수 시각화
plt.title('신도림 연령별 인구 구조')
plt.style.use('ggplot')
plt.plot(result)
plt.show()
```

## 3.개령면 지역의 인구 구조 시각화

```python
print('경상북도 김천시 개령면(4715034000)' == '개령면')

# in 연산자를 활용해서 이름 포함한 지역 찾기
print('경상북도 김천시 개령면(4715034000)' in '개령면')
print('개령면' in '경상북도 김천시 개령면(4715034000)')
```

```python
#인구 구조 결과를 담을 수 있는 리스트 선언
result = []

# 1) 파일 읽기
f = open('/content/drive/MyDrive/ABC/201902_201902_연령별인구현황_월간.csv')

data = csv.reader(f)

# 2) 데이터  확인
for row in data:
  # 3) 개령면 지역의 정보만 담기
  if '개령면' in row[0]:
    # 4) 0~99세 인구수 담기
    for i in row[3:]:
      result.append(int(i.replace(',','')))

# 5) 연령별 인구수 시각화
plt.title('개령면 연령별 인구 구조')
plt.style.use('ggplot')
plt.plot(result)
plt.show()
```

## 4.사용자 입력을 받아서 인구 구조 시각화 하기

```python
# 인구 구조 결과를 담을 수 있는 리스트 선언
result = []

# 1) 파일 읽기
f = open('/content/drive/MyDrive/ABC/201902_201902_연령별인구현황_월간.csv')

data = csv.reader(f)

# 2) 사용자 입력
region = input('인구 구조 시각화할 지역: ')

# 2) 데이터 확인
for row in data:
    # 3) 사용자가 입력한 지역의 정보만 담기
    if region in row[0]:
        # 4) 0~99세 인구수 담기
        for i in row[3:]:
            result.append(int(i.replace(',', '')))
        break

f.close()

# 5) 연령별 인구수 시각화
plt.title(f'{region} 연령별 인구 구조')
plt.style.use('ggplot')
plt.plot(result)
plt.show()
```

### 사용자 입력 (시, 동) 받아서 인구 구조 시각화

```python
#인구 구조 결과를 담을 수 있는 리스트 선언
result = []

# 1) 파일 읽기
f = open('/content/drive/MyDrive/ABC/201902_201902_연령별인구현황_월간.csv')

data = csv.reader(f)

# 2) 사용자한테 입력 받기
si = input('시(광역시) 입력하세요')
dong = input('동네명(동) 입력하세요')

# 3) 데이터  한줄씩 확인
for row in data:
  # 4) 사용자가 입력한 정보만 추출하기
  if si in row[0] and dong in row[0]:
    # 5) 0~99세 인구수 담기
    for i in row[3:]:
      result.append(int(i.replace(',','')))

f.close()

# 6) 연령별 인구수 시각화
plt.title(f'{si} {dong} 연령별 인구 구조')
plt.style.use('ggplot')
plt.plot(result)
plt.show()
```

```python
# 인구 구조 결과를 담을 수 있는 리스트 선언
result = []

# 1) 파일 읽기
f = open('/content/drive/MyDrive/ABC/201902_201902_연령별인구현황_월간.csv')

data = csv.reader(f)

# 2) 사용자한테 입력 받기
si = input('시(광역시) 입력하세요: ')
dong = input('동네명(동) 입력하세요: ')

# 3) 데이터 한 줄씩 확인
for row in data:
    # 4) 사용자가 입력한 정보만 추출하기
    if si in row[0] and dong in row[0]:
        # 5) 0~99세 인구수 담기
        for i in row[3:]:
            result.append(int(i.replace(',', '')))
        break  # 해당 지역의 데이터를 찾았으면 반복 종료

# 파일 닫기
f.close()

# 6) 연령별 인구수 시각화
plt.title(f'{si} {dong} 연령별 인구 구조')
plt.style.use('ggplot')
plt.plot(result)
plt.show()
```
