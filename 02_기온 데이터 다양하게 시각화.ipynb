## 2.서울의 월별 최고기온 분포도 그리기


*   조건 1) 사용자에게 월 1개 입력받기
*   조건 2) 입력 받은 월의 최고, 최저 기온 분포를 한번에 그리기
*   조건 3) 색을 다르게 표현하기(최저-파랑, 평균-초록, 최고-빨강)




```python
import csv
import matplotlib.pyplot as plt

f = open('/content/drive/MyDrive/ABC/seoul.csv', 'r', encoding='euc=kr')

data = csv.reader(f)
header = next(data)

# 1) 결측치를 제외한 데이터만 시각화
high = []
low = []
avg = []

# 2) 사용자에게 원하는 달 입력받기
choice_month = input('원하는 달을 선택하세요: ')

# 3) 사용자가 입력한 월의 최고, 최저, 평균 데이터 추출

for row in data:
  # 3-2) 결측치 처리
  if row[-1] != '' and row[-2] != '' and row[-3] != '' :
    # 3-3) 사용자가 입력한 월 데이터만 담기 1907-10-01 -> 08 추출 ->
    if row[0].split('-')[1] == choice_month:
      high.append(float(row[-1]))
      low.append(float(row[-2]))
      avg.append(float(row[-3]))

# 4)사용자가 입력한 월에 대한 최저, 평균, 최고 기온 분포도 그리기
plt.title('서울의 월 최저, 평균, 최고 기온 분포도')
plt.hist(high, bins=100, color='r')
plt.hist(high, bins=100, color='g')
plt.hist(high, bins=100, color='b')

plt.show()

f.close()
```

```python
import csv
import matplotlib.pyplot as plt

f = open('/content/drive/MyDrive/ABC/seoul.csv', 'r', encoding='euc-kr')

data = csv.reader(f)
header = next(data)

# 1) 결측치를 제외한 데이터만 시각화
high = []
low = []
avg = []

# 2) 사용자에게 원하는 달 입력받기
choice_month = input('원하는 월을 입력하세요. ')

# 3) 사용자가 입력한 월의 최고, 최저, 평균 데이터 추출
for row in data:
    # 3-1) 결측치 처리
    if row[-1] != '' and row[-2] != '' and row[-3] != '':
        # 3-2) 사용자가 입력한 월 데이터만 담기 1907-10-01 -> 08 추출 ->
        if row[0].split('-')[1] == choice_month:
            high.append(float(row[-1]))
            low.append(float(row[-2]))
            avg.append(float(row[-3]))

# 4) 사용자가 입력한 월에 대한 최저, 평균, 최고 기온 분포도 그리기
plt.title(f'서울의 {choice_month}월 최고, 평균, 최저 기온 분포도')

plt.hist(high, bins=100, color='red', label=f'{choice_month}월 최고 기온')
plt.hist(avg, bins=100, color='green', label=f'{choice_month}월 평균 기온')
plt.hist(low, bins=100, color='blue', label=f'{choice_month}월 최저 기온')

plt.legend(loc='upper right')
plt.show()

f.close()
```

## 3.박스 플롯을 사용하여 서울 기온을 확인하기

### 01월과 08월 최고 기온 분포도 그리기

```python
import csv
import matplotlib.pyplot as plt

f = open('/content/drive/MyDrive/ABC/seoul.csv', 'r', encoding='euc-kr')

data = csv.reader(f)
header = next(data)

# 1) 결측치를 제외한 데이터만 시각화
# 1) 1월달 최고기온, 8월달 최고 기온 달을 리스트 변수 선언
jan_temp = []
aug_temp = []

# 2) 01월과 08월의 최고기온 데이터 추출
for row in data:
    # 3-1) 결측치 처리
    if row[-1] != '':
      # 3-2) 01월달 데이터 추출
      if row[0].split('-')[1] == '01':
        jan_temp.append(float(row[-1]))

      # 3-3) 08월달 데이터 추출
      if row[0].split('-')[1] == '08':
        aug_temp.append(float(row[-1]))

# 4) 01월, 08월에 대한 최고 기온 분포도 그리기
plt.title('서울의 01월, 08월에 대한 최고 기온 분포도')
plt.hist(jan_temp, bins=100, color='b', label='01월 최고 기온')
plt.hist(aug_temp, bins=100, color='r', label='08월 평균 기온')

plt.legend()
plt.show()

f.close()
```

### 01월과 08월 최고 기온 박스플롯 그리기

```python
import csv
import matplotlib.pyplot as plt

f = open('/content/drive/MyDrive/ABC/seoul.csv', 'r', encoding='euc-kr')

data = csv.reader(f)
header = next(data)

# 1) 1월달 최고기온, 8월달 최고 기온 달을 리스트 변수 선언
jan_temp = []
aug_temp = []

# 2) 01월과 08월의 최고기온 데이터 추출
for row in data:
    # 3-1) 결측치 처리
    if row[-1] != '':
      # 3-2) 01월달 데이터 추출
      if row[0].split('-')[1] == '01':
        jan_temp.append(float(row[-1]))

      # 3-3) 08월달 데이터 추출
      if row[0].split('-')[1] == '08':
        aug_temp.append(float(row[-1]))

# 4) 01월, 08월에 대한 최고 기온 분포도 그리기
plt.title('서울의 01월, 08월에 대한 최고 기온 박스플롯')
plt.boxplot([jan_temp, aug_temp])
plt.xticks([1,2], ['01월의 최고 기온', '08월의 최고 기온'])
plt.show()

f.close()
```
