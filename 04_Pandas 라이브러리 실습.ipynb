### 01-3 파생변수 만들기

#### 변수 조합해서 파생변수 만들기

##### mpg 통합 연비 변수 만들기

```python
# 통합 연비 계산 => (도시 연비 + 고속도로 연비)/2

df_mpg['total'] = (df_mpg['cty'] + df_mpg['hwy']) / 2
df_mpg.head()
```

#### 조건을 활용해 파생 변수 만들기

##### 연비의 합격(pass), 불합격(fail) 판정 파생 변수 만들기

##### 1.합격과 불합격의 기준 정하기 = 20 기준

```python
df_mpg['total'].describe() #요약 통계량
```

```python
# 그래프로 표현
df_mpg['total'].plot.hist();
```

##### 2.합격판정 파생변수 만들기 = 20기준

```python
import numpy as np

# total 값이 20 이상이면 pass, 이하면 fail 부여
df_mpg['test'] = np.where(df_mpg['total'] >= 20, 'pass', 'fail') #np.where(조건, 만족하면 처리, 만족하지 않으면)
df_mpg.head(20)
```

##### 3.빈도표로 합격 판정 자동차 수 확인

```python
df_mpg['test'].value_counts()
```

##### 4.막대 그래프로 빈도 표현하기

```python
df_mpg['test'].value_counts().plot.bar(rot =0); #축 이름을 수평으로 만들기
```

##### 5.파이차트로 비율 표현하기

```python
import matplotlib.pyplot as plt

df_mpg['test'].value_counts().plot.pie(autopct='%.1f%%');
plt.title('pass or fail')
plt.show()
```

##02 자유자재로 데이터 가공하기

### 02-1 데이터 전처리 - 원하는 형태로 데이터 가공하기
| 메서드          | 기능              |
|---------------|-------------------|
| query()       | 행 추출           |
| df[]          | 열(변수) 추출     |

| groupby()     | 집단별로 나누기   |
| agg()         | 통계치 구하기     |
| merge()       | 데이터 합치기(열) |
| concat()      | 데이터 합치기(행) |

#### 조건에 맞는 데이터만 추출하기

| 논리 연산자 | 기능        |
|:-----------:|-------------|
| <           | 작다        |
| <=          | 작거나 같다 |
| >           | 크다        |
| >=          | 크거나 같다 |
| ==          | 같다        |
| !=          | 같지 않다   |
| \|          | 또는        |
| &           | 그리고      |
| in          | 매칭 확인   |

| 산술 연산자 | 기능            |
|:-----------:|-----------------|
| +           | 더하기          |
| -           | 빼기            |
| *           | 곱하기          |
| **          | 제곱            |
| /           | 나누기          |
| //          | 나눗셈의 몫     |
| %           | 나눗셈의 나머지 |

```python
df_mpg.info()
```

```python
# df_mpg.category의 종류를 확인
df_mpg['category'].unique()
```

```python
# df_mpg.category 2seater인 경우만 추출
df_mpg.query("category == '2seater'")
```

```python
# df_mpg.category minivan이 아닌 경우 추출
df_mpg.query("category != 'minivan'")
```

```python
# 초과(>), 미만(<), 이상(>=), 이하(<=)
# 연식이 2000을 초과한 경우만 추출
df_mpg.query('year > 2000')
```

#### 여러 조건으로 충족하는 행 추출하기

```python
# minivan 이면서 연식이 2000년 이상인 경우만 데이터 추출

df_mpg.query("category == 'minivan' & year >= 2000")
```

```python
# 차종이 minivan, 2seater, pickup 인 경우만 데이터 추출
df_mpg.query("category == 'minivan' | category == '2seater' | category == 'pickup'")
```

```python
# 차종이 minivan, 2seater, pickup 인 경우만 데이터 추출
df_mpg.query("category in ['minivan', '2seater', 'pickup']")
```

### 02-3 순서대로 정렬하기

#### 오름차순으로 정렬하기

```python
# 연식 기준으로 데이터 정렬 -> 기본 오름차순 (작은것에서 큰것으로 정렬) -> 연식이 오래된 순으로 확인 가능
df_mpg.sort_values('year')
```

#### 내림차순 정렬하기

```python
# 연식 기준으로 데이터 정렬 -> 내림차순 정렬(큰것에서 작은것으로 정렬) -> 연식이 최신순으로 확인 가능
df_mpg.sort_values('year', ascending=False)
```

```python
# 연식 기준으로 데이터 정렬 -> 내림차순 정렬(큰것에서 작은것으로 정렬) -> 연식이 최신순으로 확인 가능
df_mpg.sort_values(['year', 'category'], ascending=False)
```

```python
# 연식 기준으로 데이터 정렬 -> 연식은 최신으로, 이름은 오름차순으로 정렬
df_mpg.sort_values(['year', 'model'], ascending=[False, True])
```
